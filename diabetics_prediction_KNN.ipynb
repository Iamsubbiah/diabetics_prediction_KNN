## Values of columns like 'Glucose', Blood Pressure' cannot be accepted as zeroes because it will affect the outcome
# We can replace such values with the mean of the respective column:

```python
# Replace zeroes
zero_not_accepted = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin']
```

```python

for i in zero_not_accepted:
    # print(df[i])
    df[i]=df[i].replace(0, np.nan)
    mean=int(df[i].mean(skipna=True))
    df[i]=df[i].replace(np.nan,mean)
```

```python
# split dataset
x = df.iloc[:, 0:8]  #ilocation[all the rows(:)] , 0:8 as 9th one is result outcome column
y = df.iloc[:, 8]   #ilocation[all the rows(:)] , just 8 as its is result outcome column
x_train, x_test, y_train, y_test =train_test_split(x, y, random_state=0, test_size=0.2)
```

```python
#Feature scaling
sc=StandardScaler()
x_train =sc.fit_transform(x_train)
x_test =sc.transform(x_test)
```

```python
len(df)
```

```python
import math
math.sqrt(len(y_test))  # we dont use even numbers of neibhours as if theres eq number of both cases well get no resulkt so we take 11
```

```python
#Define the model: Init K-NN
knn=KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean') #p=2 posiblity diabrtic/not
```

```python
# Fit Model
knn.fit(x_train, y_train)
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='euclidean', metric_params=None, n_jobs=1, n_neighbors=11, p=2, weights='uniform')
```

```python
# Predict the test set results
y_pred=knn.predict(x_test)
y_pred
```

```python
# Evaluate Model
cm=confusion_matrix(y_test, y_pred) 
print (cm)
print(f1_score(y_test, y_pred))
```
